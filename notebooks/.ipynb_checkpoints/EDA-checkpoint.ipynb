# ndvi of central pixel

```python
def process_im(fid, folder='../data/raw/image_arrays_train'):
    fn = f'{folder}/{fid}.npy'
    arr = np.load(fn)
    
    bands_of_interest = ['S2_B8', 'S2_B4', 'S2_B2','S2_B3']
    values = {}
    for month in range(12):
        bns = [str(month) + '_' + b for b in bands_of_interest] # Bands of interest for this month 
        idxs = np.where(np.isin(band_names, bns)) # Index of these bands
        #vs = arr[idxs, 20, 20] # Sample the im at the center point
        #print(vs.shape)
        vs = arr[idxs, 19:22, 19:22]
        #vs = [i.mean() for i in vs]
        for bn, v in zip(bns, vs[0]):
            #print(bn)
            values[bn] = v.mean()
    return values
```

```python
# Make a new DF with the sampled values from each field 
train_high_quality = train[train['Quality'] == 3]
train_sampled = pd.DataFrame([process_im(fid) for fid in train_high_quality['Field_ID'].values])

# Add in the field ID and yield
train_sampled['Field_ID'] = train_high_quality['Field_ID'].values
train_sampled['Yield'] = train_high_quality['Yield'].values
train_sampled.head()
```

```python
def calc_vegetation_indexes(df_input):
    df = df_input.copy()
    monthes = set(col.split('_')[0] for col in df.columns if col not in ['Field_ID', 'Yield'])
    #print(monthes)
    #for month in range(12):
    for month in monthes:
        # ndvi
        tru_month = int(month)+1
        ndvi = (df[f'{month}_S2_B8'] - df[f'{month}_S2_B4']) / (df[f'{month}_S2_B8'] + df[f'{month}_S2_B4']) 
        df[f'{tru_month}_ndvi'] = ndvi
        #evi
        evi = 2.5 * (df[f'{month}_S2_B8'] - df[f'{month}_S2_B4']) / \
                             (df[f'{month}_S2_B8'] + 6 * df[f'{month}_S2_B4'] - 7.5 * df[f'{month}_S2_B2'] + 1) 

        df[f'{tru_month}_evi'] = evi
        
        grvi = df[f'{month}_S2_B8'] / df[f'{month}_S2_B3']
        df[f'{tru_month}_grvi'] = grvi
        
        gndvi = (df[f'{month}_S2_B8'] - df[f'{month}_S2_B3']) / (df[f'{month}_S2_B8'] + df[f'{month}_S2_B3'])
        df[f'{tru_month}_gndvi'] = gndvi
        
        alpha = 0.1
        wrdvi = (alpha*df[f'{month}_S2_B8'] - df[f'{month}_S2_B4']) / (alpha*df[f'{month}_S2_B8'] + df[f'{month}_S2_B4']) 
        df[f'{tru_month}_wrdvi'] = wrdvi
    return df
```

```python
df_with_indexes = calc_vegetation_indexes(train_sampled)
```

0_CLIM_pet - Reference evapotranspiration(ASCE grass/alfalfa)

0_CLIM_pr - Precipitation accumulation -  

0_CLIM_ro - Runoff

0_CLIM_soil - Soil moisture

0_CLIM_srad - Downward solar radiation

0_CLIM_swe - Snow water equivalent

0_CLIM_tmmn- temperature min 

0_CLIM_tmmx - temperature max 

0_CLIM_vap - Vapor Pressure

0_CLIM_vpd - Vapor Pressure Deficit

```python
weather_expl = {
'CLIM_aet' : 'Actual evapotranspiration - ',
'CLIM_def'  : 'Climate water deficit ',
'CLIM_pdsi' : 'Palmer drought severity index (PDSI)',
'CLIM_pet'  : 'Reference evapotranspiration(ASCE grass/alfalfa)',
'CLIM_pr'   : 'Precipitation accumulation -  ',
'CLIM_ro'   : 'Runoff',
'CLIM_soil' : 'Soil moisture',
'CLIM_srad' : 'Downward solar radiation',
'CLIM_swe'  : 'Snow water equivalent',
'CLIM_tmmn' : 'temperature min ',
'CLIM_tmmx' : 'temperature max ',
'CLIM_vap'  : 'Vapor Pressure',
'CLIM_vpd'  : 'Vapor Pressure Deficit',
'CLIM_vs'   : 'Wind Speed (10-m)'}
```

# prepare ndvi image

```python
def process_im(fid, folder='../data/raw/image_arrays_train'):
    fn = f'{folder}/{fid}.npy'
    arr = np.load(fn).astype(float)
    
    bands_of_interest = ['S2_B8', 'S2_B4']
    nir_idx = 303
    r_idx = 307
    ndvi_maps = []
    for month in range(12):
        bns = [str(month) + '_' + b for b in bands_of_interest]
        idxs = np.where(np.isin(band_names, bns))
        #print(month, bns, idxs[0], )
        
        ndvi_map = (arr[idxs[0][0]] - arr[idxs[0][1]]) / (arr[idxs[0][0]] + arr[idxs[0][1]]) 
        ndvi_maps.append(ndvi_map)
    return ndvi_maps
```

```python
def process_im_for_evi(fid, folder='../data/raw/image_arrays_train'):
    fn = f'{folder}/{fid}.npy'
    arr = np.load(fn).astype(float)
    
    bands_of_interest = ['S2_B8', 'S2_B4', 'S2_B2']
    nir_idx = 303
    r_idx = 307
    ndvi_maps = []
    for month in range(12):
        bns = [str(month) + '_' + b for b in bands_of_interest]
        idxs = np.where(np.isin(band_names, bns))
        #print(month, bns, idxs[0], )
        
        ndvi_map = 2.5*(arr[idxs[0][0]] - arr[idxs[0][1]]) / (arr[idxs[0][0]] + 6* arr[idxs[0][1]] +1 + 7.5 * arr[idxs[0][2]]) 
        ndvi_maps.append(ndvi_map)
    return ndvi_maps
```

```python
fields_ndvi_map = {}
for filed in df_with_indexes['Field_ID']:
    fields_ndvi_map[filed] = process_im(filed)
```

```python
fields_evi_map = {}
for filed in df_with_indexes['Field_ID']:
    fields_evi_map[filed] = process_im_for_evi(filed)
```

```python
def get_map_for_field(fid, folder='../data/raw/image_arrays_train'):
    fn = f'{folder}/{fid}.npy'
    arr = np.load(fn).astype(float)
    conv = arr[3][:40, :40]
    #print(conv.shape)

    return conv
```

```python
maps = [get_map_for_field(fid, folder='../data/raw/image_arrays_train/') for fid in train['Field_ID'].values[:]]
```

```python
all_maps = np.array(maps)
```

```python
corr_m = np.zeros((40, 40))

true = train['Yield'].values
for i in range(40):
    for j in range(40):
        vals = all_maps[:, i, j]
        corr = np.corrcoef(vals, true)
        corr_m[i, j] = corr[0, 1]
```
